```python
df_train.corr().spread_Won.sort_values()
```

```python
df_train_y = df_train['spread_Won']
df_train_x = df_train.drop(['spread_Won', 'spread_Lost', 'spread_Push', 'OU_Over', 'OU_Push', 'OU_Under'], axis = 'columns')
df_test_y = df_test['spread_Won']
df_test_x = df_test.drop(['spread_Won', 'spread_Lost', 'spread_Push', 'OU_Over', 'OU_Push', 'OU_Under'], axis = 'columns')
```

```python
#We will use a MinMax Scaler to transform our data to a [0, 1] scale
scaler = MinMaxScaler()
df_train_x = scaler.fit_transform(df_train_x)
df_train_x = pd.DataFrame(df_train_x, columns = df_test_x.columns)
df_test_x = scaler.fit_transform(df_test_x)
df_test_x = pd.DataFrame(df_test_x, columns = df_train_x.columns)
```

```python
#Chi square feature selection is used to select the 50 best features. This method keeps the features with the 50 highest 
#chi square test statistic value in our model
chi2_features = SelectKBest(chi2, k = 50)
x_50best_features = chi2_features.fit_transform(df_train_x, df_train_y)
features = chi2_features.get_support(indices=True)
features
features_df_new = df_train_x.iloc[:,features]
features_df_new
```

```python
#correlation table of our 50 selected features with our predicted variable
pd.concat([features_df_new, df_train_y], axis = 'columns').corr()
```

```python
# random search logistic regression model on the sonar dataset
from scipy.stats import loguniform
from pandas import read_csv
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import RandomizedSearchCV
# define model
model = LogisticRegression()
# define evaluation
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
# define search space
space = dict()
space['solver'] = ['newton-cg', 'lbfgs', 'liblinear']
space['penalty'] = ['none', 'l1', 'l2']
space['C'] = loguniform(1e-5, 10)
# define search
search = RandomizedSearchCV(model, space, n_iter=100, scoring='accuracy', n_jobs=-1, cv=cv, random_state=1)
# execute search
result = search.fit(features_df_new, df_train_y)
# summarize result
print('Best Score: %s' % result.best_score_)
print('Best Hyperparameters: %s' % result.best_params_)
```

```python
#we will now evaluate the performance of our best model
yhat = result.best_estimator_.predict(df_test_x.iloc[:,features])
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print("=== Confusion Matrix ===")
print(confusion_matrix(df_test_y, yhat))
print('\n')
print("=== Classification Report ===")
print(classification_report(df_test_y, yhat))
```
